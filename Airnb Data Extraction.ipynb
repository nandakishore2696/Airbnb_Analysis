## Handling Null Values

```python
df.isna().sum()
```

```python
# replace review_rating and cleaning_fee with mean rating
df['review_rating'].fillna(df['review_rating'].mean(),inplace=True)
df['cleaning_fee'].fillna(df['cleaning_fee'].mean(),inplace=True)
```

```python
df['bedrooms'].fillna(df['bedrooms'].median(), inplace=True)
```

```python
# replace beds and bathrooms with number of bedrooms
df['beds'].fillna(df['bedrooms'], inplace=True)
df['bathrooms'].fillna(df['bedrooms'], inplace=True)
```

```python
# As the name doesn't matter, replace 
df['property_name'].fillna('abc', inplace=True)
```

```python
df.to_csv("Airbnb_cleaned_data.csv")
```
